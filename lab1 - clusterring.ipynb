```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Reading a CSV in Pandas
df = pd.read_csv('salary for kmeans.csv')
df
```

```python
#Scatter plot
sns.scatterplot(df["YearsExperience"],df['Salary'])
```

```python
from sklearn.cluster import KMeans

KM_Model = KMeans(n_clusters = 3 )
cluster_predict_old = KM_Model.fit_predict(df[['YearsExperience','Salary']])

cluster_predict_old
```

```python
df['old_cluster'] = cluster_predict_old
df
```

```python
df0 = df[df['old_cluster'] == 0]
df1 = df[df['old_cluster'] == 1]
df2 = df[df['old_cluster'] == 2]
```

```python
# Because of wrong clusturing we will downsacle the Age and Income($) value
# We will convert all values in range of 0-1
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()

# Transform YearsExperience
Scaler.fit(df[['YearsExperience']])
df['YearsExperience'] = Scaler.transform(df[['YearsExperience']])

# Transform Salary
Scaler.fit(df[['Salary']])
df['Salary'] = Scaler.transform(df[['Salary']])
```

```python
df
```

```python
#Scatter plot
sns.scatterplot(df["YearsExperience"],df['Salary'])
```

```python
from sklearn.cluster import KMeans

KM_Model = KMeans(n_clusters = 3 )
cluster_predict_new = KM_Model.fit_predict(df[['YearsExperience','Salary']])

cluster_predict_new
```

```python
df['new_cluster'] = cluster_predict_new
df
```

```python
df0 = df[df['new_cluster'] == 0] 
df1 = df[df['new_cluster'] == 1]
df2 = df[df['new_cluster'] == 2]
```

```python
df0
```

```python
df1
```

```python
df2
```

```python
sns.scatterplot(df0["YearsExperience"],df0['Salary'],color='red')
sns.scatterplot(df1["YearsExperience"],df1['Salary'],color='blue')
sns.scatterplot(df2["YearsExperience"],df2['Salary'],color='green')
```

```python
# Find K value in Kmeans :- we will used elbow method

SSE = []

for k in range(1,16):
    KM_Model = KMeans(n_clusters = k )
    KM_Model.fit(df[['YearsExperience','Salary']])
    SSE.append(KM_Model.inertia_)
    
SSE
```

```python
# Plot between Acc and K value.
# X_Axis = K value
# Y_Axis = SSE
plt.plot(range(1,16),SSE,'*-')
plt.xlabel("K_Value")
plt.ylabel("Acc")
```

```python
#From above graph we come to know that perfect value for K is 4 
```
